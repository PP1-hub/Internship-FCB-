```python
path = r"C:\Users\baiba\OneDrive\Desktop\\"
moveto = r"C:\Users\baiba\OneDrive\Desktop\intern_files"
files = os.listdir(path)
files.sort()
for f in files:
    src = path + f
    dst = moveto + f
    shutil.move(src,dst)
    if not f.startswith('.'):
        if not f.endswith('.rar'):
            for ff in os.listdir(path + f + '\ '):
                src1 = path + f + '\ ' + ff
                dst1 = moveto + ff
                if(len(os.listdir(path + f + '\ ')) == 0):
                    shutil.move(src1, dst1)
                else:
                    if not ff.startswith('.'):
                        for fff in os.listdir(path + f + '\ ' + ff + '\ '):
                            src2 = path + f + '\ ' + ff + '\ ' + fff
                            dst2 = moveto + fff
                            shutil.move(src2, dst2)
```
